```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('after_all_preprocess.csv')
df.head(5)
```

```python
df['clean_sentence'] = df['clean_sentence'].str.replace("''","")
```

```python
count = []
for i in df['clean_sentence']:
  count.append(i)
```

```python
count_words = ' '.join([str(elem) for elem in count])
print(count_words)
```

```python
No_of_character = len(count_words)
No_of_words = len(count_words.split())
print(No_of_character)
print(No_of_words)
```

```python
import nltk
from nltk.probability import FreqDist
frequency_dist = nltk.FreqDist(count_words.split())
frequency_dist
```

```python
sorted_frequency_dist =sorted(frequency_dist,key=frequency_dist.__getitem__, reverse=True)
```

```python
sorted_frequency_dist
```

```python
large_words = dict([(k,v) for k,v in frequency_dist.items() if len(k)>3])
frequency_dist = nltk.FreqDist(large_words)
frequency_dist.plot(20,cumulative=False)
```

```python
from wordcloud import WordCloud
wcloud = WordCloud().generate_from_frequencies(frequency_dist)
#plotting the wordcloud
import matplotlib.pyplot as plt
plt.imshow(wcloud, interpolation='bilinear')
```
